```
import bs4
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```
#Coletando as informações

def fundamentusdata ():

  pd.set_option('display.max_colwidth', 1000)

  link = requests.get("https://www.fundamentus.com.br/resultado.php")

  soup = bs4.BeautifulSoup(link.text,"html.parser")
  table = str(soup.find(name="table"))

  data = pd.read_html(table, decimal = ",", thousands=".")[0]

  #Renomeando as colunas
  data.columns = ["papel","cotacao","pl","pvp","psr","divyield","pativo","pcapgiro","pebit","pativcircliq","evebit","evebitda","mrgebit","mrgliq","liqcorr","roic","roe","liq2meses","patrimliq","divbrutpatrim","crescrec5a" ]


  #Tratando as informações
  #Removendo o %
  data.replace(to_replace = "%",value = "", inplace = True, regex=True)

  #Removendo o "." dos milhares e trocando os decimais
  data["divyield"] = data["divyield"].str.replace(",", ".") 
  data["mrgebit"] = data["mrgebit"].str.replace(".", "").str.replace(",", ".")
  data["mrgliq"] = data["mrgliq"].str.replace(".", "").str.replace(",", ".")
  data["roic"] = data["roic"].str.replace(".", "").str.replace(",", ".")
  data["roe"] = data["roe"].str.replace(".", "").str.replace(",", ".")
  data["crescrec5a"] = data["crescrec5a"].str.replace(".", "").str.replace(",", ".")


  #Convertendo em númerico (float)
  data["divyield"] = pd.to_numeric(data["divyield"], errors='coerce')
  data["mrgebit"] = pd.to_numeric(data["mrgebit"], errors='coerce')
  data["mrgliq"] = pd.to_numeric(data["mrgliq"], errors='coerce')
  data["roic"] = pd.to_numeric(data["roic"], errors='coerce')
  data["roe"] = pd.to_numeric(data["roe"], errors='coerce')
  data["crescrec5a"] = pd.to_numeric(data["mrgebit"], errors='coerce')


  #Removendo ações sem negociação
  indexzero = data[data["liq2meses"] == 0 ].index
  data.drop(indexzero , inplace=True)

  return data.sort_values("cotacao", ascending= False)



```
